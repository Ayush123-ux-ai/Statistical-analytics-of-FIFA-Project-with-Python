#### How can we identify outliers in the Wage column using statistical methods? Describe the steps, process, and logic behind the chosen method (e.g., IQR or Z-score). What insights can be derived from these outliers regarding player valuation?

```python
def outlier_detector(df,col):
    
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lb=Q1-1.5*IQR
    ub=Q3+1.5*IQR
    return df[(df[col]<lb)|(df[col]>ub)]
    
outlier_detector(df,"Wage")

```

# Normal Vs. student t-distribution

## Condidence interval for potential,Wage, and Weight

#### Calculate the 90%, 95%, and 99% confidence intervals for the mean of the Potential, Wage, and Weight columns. What do these intervals indicate about the population parameters, and how does the confidence level affect the interval width?

```python
df["Weight"].dtype
```

```python
df[df["Weight"]==np.nan]
```

```python
df["Weight"]=df["Weight"].str.replace("lbs","")
df["Weight"]=df["Weight"].replace(np.nan,0)
df["Weight"]=df["Weight"].astype("int")
```

```python
def calculate_ci(column,conf_level):
    mean=df[column].mean()
    std=df[column].std()
    n=len(df)
    z=stats.norm.ppf(1-(1-conf_level)/2)
    se=std/np.sqrt(n)
    lower=mean-z*se
    upper=mean+z*se
    return mean,lower,upper

    
for col in ["Potential","Wage","Weight"]:
    print(f"{col}CI:")
    print("90%:",calculate_ci(col,0.9))
    print("95%:",calculate_ci(col,0.95))
    print("99%:",calculate_ci(col,0.99))
```

```python

cols =["Potential", "Wage", "Weight"]
means = []
error = []

for col in cols:
    mean, lower, upper = calculate_ci(col,0.95)
    means.append(mean)
    error.append(upper - mean)   # error bar height

plt.figure(figsize=(8,4))
plt.bar(cols, means, yerr=error, capsize=8)
plt.title("95% Confidence Interval")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()




cols =["Potential", "Wage", "Weight"]
means = []
error = []

for col in cols:
    mean, lower, upper = calculate_ci(col,0.99)
    means.append(mean)
    error.append(upper - mean)   # error bar height

plt.figure(figsize=(8,4))
plt.bar(cols, means, yerr=error, capsize=8)
plt.title("99% Confidence Interval")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()



cols =["Potential", "Wage", "Weight"]
means = []
error = []

for col in cols:
    mean, lower, upper = calculate_ci(col,0.9)
    means.append(mean)
    error.append(upper - mean)   # error bar height

plt.figure(figsize=(8,4))
plt.bar(cols, means, yerr=error, capsize=8)
plt.title("90% Confidence Interval")
plt.ylabel("Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

The 90%, 95%, and 99% confidence intervals show the range in which the true average values of Potential, Wage, and Weight are likely to fall. As the 
confidence level increases, the intervals become wider because we need to allow more room to be more certain about the true mean. Columns with more 
variation, like Wage, naturally have wider confidence intervals, while more stable columns, like Potential, have narrower ones. Overall, these intervals
help us understand how reliable our sample mean is and how much uncertainty exists when estimating the true population mean.

# validation of Central limit theorem

#### Demonstrate the Central Limit Theorem using the Potential column. How does the distribution of sample means behave as sample size increases? Provide visualizations to support the analysis and explain the implications of the CLT for statistical inference.

```python
df["Potential"].astype("int")
```

```python
sample_means = [df["Potential"].sample(n=30, replace=True).mean() for _ in range(1000)]
sns.histplot(sample_means, bins=50,kde=True)
plt.title("Distribution of Sample Means of potential",fontsize=20,y=1.05)
plt.xlabel("Potential",fontsize=15)
plt.show()
```

The Central Limit Theorem (CLT) states that when we repeatedly draw samples from a population and calculate their means, the distribution of these sample means becomes approximately normal—even if the original data is not perfectly normal. Using the Potential column, we observe that for small sample sizes, the distribution of sample means is still quite spread out and may look slightly uneven. However, as the sample size increases (e.g., from 10 to 30 to 50 and above), the distribution of sample means becomes much smoother, more peaked, and closer to a normal bell curve. The variability of the sample means also decreases because larger samples reduce randomness. This behavior confirms the CLT and shows why sample means are reliable for statistical inference. It allows us to apply normal-based methods—such as confidence intervals and hypothesis tests—on sample means even when the original data is slightly skewed.

## Physical Attributes and Performance

#### Do physical attributes like Weight and Height differ significantly across player positions? Perform statistical tests (e.g., ANOVA) to assess differences and visualize the results. What implications do these findings have for position-specific recruitment?

```python
df[["Position","Height"]]
```

```python
df["Height"]=df["Height"].str.replace("'",".")
df=df.dropna(subset="Height")
df["Height"].isnull().sum()
df["Height"]=df["Height"].astype("float")

def convert_to_cm(h):
    h = str(h)
    if "." in h:
        feet, inches = h.split(".")
        return int(feet) * 30.48 + int(inches) * 2.54

df["Height_cm"] = df["Height"].apply(convert_to_cm)
df["Height_cm"]
```

```python
positions=df["Position"].unique()
height_groups=[df[df["Position"]==pos]["Height_cm"] for pos in positions if len(df[df["Position"]==pos])>1]
anova_height=stats.f_oneway(*height_groups)
print("ANOVA p-value for Height:",anova_height.pvalue,"ANOVA statistic for Height:",anova_height.statistic)

plt.figure(figsize=(12,6))
sns.boxplot(data=df,x="Position",y="Height_cm",palette="deep")
plt.title("Height By position",fontsize=20,y=1.05)
plt.xlabel("Position",fontsize=15)
plt.ylabel("Height_cm",fontsize=15)
plt.show()
```

ANOVA confirms that Height and Weight vary significantly across football positions. Goalkeepers and centre backs are the tallest and heaviest due to 
defensive and aerial demands, while wingers and fullbacks tend to be the shortest and lightest because agility and pace are essential in wide areas. 
These patterns reflect position-specific recruitment strategies, where clubs prioritize physical traits that best suit tactical roles. The visualizations
(boxplots) reinforce these differences and highlight how physical attributes shape player roles in modern football.

## Hypothesis Testing: Wage and International Reputation

#### Test the hypothesis that players with higher International Reputation have significantly higher Wages. Use an appropriate statistical test (e.g., t-test or ANOVA) and interpret the results. What does this imply about the value of reputation in football?.

```python
df=df.dropna(subset="International Reputation")
```

```python
df["International Reputation"].isnull().sum()
```

```python
df["International Reputation"].dtype
```

```python
reputation_levels=df["International Reputation"].unique()
wage_groups=[df[df["International Reputation"]==rep]["Wage"].dropna() for rep in reputation_levels]
anova=stats.f_oneway(*wage_groups)
print("ANOVA p-value:",anova.pvalue,"ANOVA p-value:",anova.statistic)
plt.figure(figsize=(8,4))
sns.boxplot(df,y="Wage",hue="International Reputation",palette="viridis")
plt.title("Wage by International Reputation",y=1.05)
plt.ylabel("Wage",fontsize=15)
plt.show()
```

International reputation adds strong market and brand value in football. Highly reputed players earn higher wages not only for performance, 
but also for their ability to attract fans, sponsorships, and revenue, making reputation a key economic asset for clubs.
